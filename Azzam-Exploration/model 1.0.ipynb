Import modul

```python
# Loading the required python modules
import pytesseract # this is tesseract module
import matplotlib.pyplot as plt
import cv2 # this is opencv module
import glob
import os
import pandas as pd
```

Ambil data dari directory gambar dan csv

```python
# specify path to the license plate images folder as shown below
path_for_license_plates = os.getcwd() + "/data/*.png"
predicted_license_plates = []

# read csv file and read actual license
labels_csv = pd.read_csv('./data/DataTrain.csv', delimiter=';')
list_license_plates = list(labels_csv['Vehicleregistrationplate'])
list_file_name = list(labels_csv['NameofFile'])
```

Prediksi gambar dengan TesseractOCR

```python
for i in range (1, len(list_license_plates) + 1):
      
    '''
    Read each license plate image file using openCV
    '''
    img_path = "./data/DataTrain" + str(i) + ".png"
    img = cv2.imread(img_path)
      
    '''
    We then pass each license plate image file
    to the Tesseract OCR engine using the Python library 
    wrapper for it. We get back predicted_result for 
    license plate. We append the predicted_result in a
    list and compare it with the original the license plate
    '''
    # tesseract
    pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'
    predicted_result = pytesseract.image_to_string(img, lang ='eng',
    config ='--oem 3 --psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789')
      
    filter_predicted_result = "".join(predicted_result.split()).replace(":", "").replace("-", "")
    predicted_license_plates.append(filter_predicted_result)
    print("Predicted DataTrain" + str(i) + ".png")
```

Fungsi buat ngeprint hasil prediksi

```python
def calculate_predicted_accuracy(names, actual_list, predicted_list):
    for name, actual_plate, predict_plate in zip(names, actual_list, predicted_list):
        accuracy = "0 %"
        num_matches = 0
        if actual_plate == predict_plate:
            accuracy = "100 %"
        else:
            if len(actual_plate) == len(predict_plate):
                for a, p in zip(actual_plate, predict_plate):
                    if a == p:
                        num_matches += 1
                accuracy = str(round((num_matches / len(actual_plate)), 2) * 100)
                accuracy += "%"
        print("{0:25}{1:25}{2:30}{3:25}".format(name, actual_plate, predict_plate, accuracy))
```

Print Hasil Prediksi

```python
print("File Name", "\t\t", "Actual License Plate", "\t", "Predicted License Plate", "\t", "Accuracy")
print("--------------------", "\t", "--------------------", "\t", "-----------------------", "\t", "--------")

calculate_predicted_accuracy(list_file_name, list_license_plates, predicted_license_plates)
```

```python
grayscale_resize_test_license_plate = cv2.cvtColor(
    resize_test_license_plate, cv2.COLOR_BGR2GRAY)
```

```python
gaussian_blur_license_plate = cv2.GaussianBlur(
    grayscale_resize_test_license_plate, (5, 5), 0)
```
